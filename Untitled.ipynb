```python
import pandas as pd
```

```python
covid_train= pd.read_csv('/home/mukund/Documents/COVID/COVID-CT/Data-split/COVID/trainCT_COVID.txt')
healthy_train= pd.read_csv('/home/mukund/Documents/COVID/COVID-CT/Data-split/NonCOVID/trainCT_NonCOVID.txt')
```

```python
print(healthy_train)
```

```python
healthy_labels= np.zeros((len(healthy_train), 1))
print(np.shape(healthy_labels))
print(healthy_labels)
```

```python
print(covid_train)
```

```python
covid_labels= np.ones((len(covid_train), 1))
print(np.shape(covid_labels))
print(covid_labels)
```

```python
covid_train_new= np.asarray(covid_train)
print(np.shape(covid_train_new))
```

```python
covid_train_new= np.concatenate((covid_train_new, covid_labels), axis=1)
print(np.shape(covid_train_new))
print(covid_train_new)
```

```python
covid_train_new[0][1]
```

```python
healthy_train_new= np.asarray(healthy_train)
healthy_train_new= np.concatenate((healthy_train_new, healthy_labels), axis=1)
print(np.shape(healthy_train_new))
print(healthy_train_new)
```

```python
train_data= np.concatenate((covid_train_new, healthy_train_new), axis=0)
print(np.shape(train_data))
```

```python
print(train_data)
```

```python
np.save('train_list.txt', train_data)
```
